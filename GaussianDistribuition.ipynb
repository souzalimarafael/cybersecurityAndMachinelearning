```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados com distribuição normal
mu, sigma = 0, 0.1 # média e desvio padrão
s = np.random.normal(mu, sigma, 1000)

# Plotar histograma com a distribuição
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')

# Adicionar título ao gráfico
plt.title('Gaussian distribution', fontsize=24)

# Adicionar texto dentro do gráfico
plt.annotate('mean', xy=(mu, 0), xytext=(mu - 0.04, 0.15), fontsize=20)
plt.annotate('lower tail', xy=(-0.3, 0.2), xytext=(-0.34, 0.3), fontsize=20)
plt.annotate('upper tail', xy=(0.3, 0.2), xytext=(0.24, 0.3), fontsize=20)

# Adicionar rótulos dos eixos x e y
plt.xlabel('Value', fontsize=20)
plt.ylabel('Density', fontsize=20)

# Aumentar a proporção do gráfico
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

# Aumentar o tamanho da fonte das palavras e números
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

# Aumentar o tamanho da fonte dos números dos índices na lateral e na horizontal
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
