```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo direcionado
G = nx.DiGraph()

# Adicionando nós
G.add_nodes_from(['IP de Origem', 'Tamanho do Pacote', 'Número de Pacotes', 'Protocolo', 'Entropy', 'Camada Oculta 1', 
                  'Camada Oculta 2', 'Camada Oculta 3', 'Saída'])

# Adicionando arestas
G.add_edges_from([('IP de Origem', 'Camada Oculta 1'), ('Tamanho do Pacote', 'Camada Oculta 1'),
                  ('Número de Pacotes', 'Camada Oculta 1'), ('Protocolo', 'Camada Oculta 1'),
                  ('Entropy', 'Camada Oculta 1'), ('Camada Oculta 1', 'Camada Oculta 2'),
                  ('Camada Oculta 1', 'Camada Oculta 3'), ('Camada Oculta 2', 'Saída'),
                  ('Camada Oculta 3', 'Saída')])

# Definindo as posições dos nós
pos = {'IP de Origem': (1, 4), 'Tamanho do Pacote': (2, 4), 'Número de Pacotes': (3, 4), 'Protocolo': (4, 4), 'Entropy': (5, 4),
       'Camada Oculta 1': (3, 3), 'Camada Oculta 2': (4, 2), 'Camada Oculta 3': (2, 2), 'Saída': (3, 1)}

# Definindo as cores das colunas
column_colors = {0: '#FFD700', 1: '#FFA500', 2: '#FF8C00', 3: '#FF4500', 4: '#FF0000'}

# Criando o plot
plt.figure(figsize=(10, 5))
nx.draw(G, pos=pos, with_labels=True, node_size=1000, node_color=[column_colors[pos[node][1]] for node in G.nodes()],
        edgecolors='black', font_size=12, font_weight='bold', arrowsize=20, arrowstyle='->', width=2)

plt.title('Fluxograma de Rede Neural para Detecção de Botnets', fontsize=16)
plt.show()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model

# Criando o modelo da rede neural
model = Sequential()
model.add(Dense(16, input_dim=5, activation='relu'))  # Camada de entrada com 5 nós
model.add(Dense(8, activation='relu'))  # Primeira camada oculta com 8 nós
model.add(Dense(4, activation='relu'))  # Segunda camada oculta com 4 nós
model.add(Dense(1, activation='sigmoid'))  # Camada de saída com 1 nó (detecção binária)

# Compilando o modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Plotando a arquitetura da rede neural
plot_model(model, to_file='botnet_detection_architecture.png', show_shapes=True, show_layer_names=True)
```
